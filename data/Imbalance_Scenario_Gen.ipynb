## System Imbalance Scenarios

The system imbalance is modeled as a Bernoulli random variable:

$$
SI_{t,s} \sim \text{Bernoulli}(p)
$$

where:
- $SI = 1$ indicates a system deficit,
- $SI = 0$ indicates a system surplus.

Four independent imbalance scenarios are generated over a 24-hour horizon. These scenarios are later combined with wind and price scenarios to form the full stochastic scenario set.

```python
# ============================================
# STEP 3 — RANDOM IMBALANCE SCENARIOS
# ============================================

import numpy as np
import pandas as pd
import os

# ----------------------------
# PARAMETERS
# ----------------------------
T = 24       # hours
N_s = 4      # number of imbalance scenarios
p = 0.5      # probability of deficit

np.random.seed(42)  # reproducibility

# ----------------------------
# GENERATE SCENARIOS
# ----------------------------
# 1 = deficit, 0 = surplus

SI = np.random.binomial(1, p, size=(T, N_s))

# ----------------------------
# CREATE DATAFRAME
# ----------------------------
scenario_names = [f"S{i+1}" for i in range(N_s)]

imbalance_df = pd.DataFrame(SI, columns=scenario_names)

imbalance_df.insert(0, "Hour", np.arange(T))

print(imbalance_df.head())

# ----------------------------
# SAVE CSV
# ----------------------------
os.makedirs("processed", exist_ok=True)

imbalance_df.to_csv(
    "processed/imbalance_scenarios_random.csv",
    index=False
)

print("Random imbalance scenarios saved!")
```
